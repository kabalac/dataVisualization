```python
#importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading the excel data using pandas library
data = pd.read_excel("MassShootings Database19822022.xlsx")
```

```python
# grouping total victims by year
total_victims_year = data.groupby('year').sum()
print('Total Victims = ' ,total_victims_year['total_victims'].sum())
total_victims_year.sort_values(by = 'total_victims', ascending = False)
```

```python
total_victims_year.head()
```

```python
total_victims_year.tail()
```

```python
# grouping total victims by State
total_victims_state = data.groupby('State').sum()
print('Total Victims = ' ,total_victims_state['total_victims'].sum())
total_victims_state.sort_values(by = 'total_victims', ascending = False)
```

```python
# grouping total victims by City
total_victims_city = data.groupby('City').sum()
print('Total Victims = ' ,total_victims_city['total_victims'].sum())
total_victims_city.sort_values(by = 'total_victims', ascending = False)
```

```python
# grouping total victims by common combined location
total_victims_location = data.groupby('location').sum()
print('Total Victims = ' ,total_victims_location['total_victims'].sum())
total_victims_location.sort_values(by = 'total_victims', ascending = False)
```

```python
total_victims_location.head()
```

```python
total_victims_location.tail()
```

```python
total_victims_place = data.groupby('event_place').sum()
print('Total Victims = ' ,total_victims_place['total_victims'].sum())
total_victims_place.sort_values(by = 'total_victims', ascending = False)
```

```python
# Visualizing the total_victims with the corresponding states in a bar chart
plt.subplots(figsize = (15, 6))
cr = total_victims_state['total_victims'].sort_values(ascending = False)
ax = cr.plot.bar()
ax.set_xlabel('State')
ax.set_ylabel('Total Number = Dead + Injured')
ax.set_title('Statewise total Reported cases')
plt.show()
print(cr)
```

```python
# Visualizing the total_victims with the corresponding states in a pie chart
plt.subplots(figsize = (15, 6))
cr = total_victims_state['total_victims'].sort_values(ascending = False)
ax = cr.plot.pie()
ax.set_xlabel('State')
ax.set_ylabel('Total Number = Dead + Injured')
ax.set_title('Statewise total Reported cases')
plt.show()
print(cr)
```

```python
# Visualizing the total_victims with the corresponding place in a bar chart
plt.subplots(figsize = (15, 6))
cr1 = total_victims_place['total_victims'].sort_values(ascending = False)
ax = cr1.plot.bar()
ax.set_xlabel('State')
ax.set_ylabel('Total Number = Dead + Injured')
ax.set_title('Statewise total Reported cases')
plt.show()
print(cr1)
```

```python
# Visualizing the total_victims with the corresponding place in a pie chart
plt.subplots(figsize = (15, 6))
cr1 = total_victims_place['total_victims'].sort_values(ascending = False)
ax = cr1.plot.pie()
ax.set_xlabel('State')
ax.set_ylabel('Total Number = Dead + Injured')
ax.set_title('Statewise total Reported cases')
plt.show()
print(cr1)
```

```python
import squarify
y = data['total_victims'].value_counts().head(25)
plt.rcParams['figure.figsize'] = (15, 15)
plt.style.use('fivethirtyeight')
color = plt.cm.magma(np.linspace(0, 1, 15))
squarify.plot(sizes = y.values, label = y.index, alpha=.8, color = color, pad = True)
plt.title('Tree Map for Top Crime', fontsize = 35)
plt.axis('off')
plt.show()
```

```python
plt.style.use('seaborn')
data['event_place'].value_counts().head(11).plot.pie(figsize = (15, 8))
plt.title('Crime Count Event Places',fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(20,6))
ax = sns.boxplot(x='State',y='total_victims',data=data)
plt.xticks(rotation=90)
ax.set_title("Statewise Crime Distribution")
```

```python
plt.figure(figsize=(20,6))
ax = sns.boxplot(x='event_place',y='total_victims',data=data)
plt.xticks(rotation=90)
ax.set_title("Statewise Crime Distribution")
```
